```python
import pickle
import warnings
import numpy as np
import pandas as pd
from copy import deepcopy
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

warnings.filterwarnings('ignore')
# pd.set_option('display.max_colwidth', 0)
```

```python
data = 'synthetic'
base_model = 'lr'

with open(f'../results/params_effect_{base_model}_{data}.pkl', 'rb') as f:
    results = pickle.load(f)
```

```python
results_opt = results[(results.id == 'OPT')]
idx = np.argmax(results_opt['Mean M1 Validity'])

alpha = results_opt['alpha'].iloc[idx]
lamb = results_opt['lambda'].iloc[idx]

results[(results['alpha'] == alpha) & (results['lambda'] == lamb)].round(2).T
```

Effect of $\lambda$

```python
Jcolor = {'OPT': '#636EFA', 'ROAR': '#EF553B'}
Vcolor = {'OPT': '#040e89', 'ROAR': '#c12910'}
Ccolor = {'OPT': '#7881fb', 'ROAR': '#f48774'}
font_family = 'Times New Roman'
font_color = 'black'
width, height = 720, 540

data_names = {
    'correction': 'German Credit',
    'synthetic': 'Synthetic',
    'temporal': 'Small Business'
}
# alpha = 0.1
curr_res = results[results['alpha'] == 0.5]

figs = [go.Figure() for _ in range(3)]
for id in results.id.unique():
    temp = curr_res[(curr_res.id == id)]
    
    figs[0].add_trace(go.Scatter(
        x=temp['lambda'],
        y=temp['Mean WC Validity'].round(2),
        error_y=dict(array=temp['Std WC Validity'].round(2)),
        marker=dict(color=Jcolor[id], symbol='circle' if id=='OPT' else 'circle-open', size=5),
        mode='markers',
        name=f'{"Alg 1" if id == "OPT" else "ROAR"}',
    ))
    
    figs[1].add_trace(go.Scatter(
        x=temp['lambda'],
        y=temp['Mean M1 Validity'].round(2),
        error_y=dict(array=temp['Std M1 Validity'].round(2)),
        marker=dict(color=Jcolor[id], symbol='circle' if id=='OPT' else 'circle-open', size=5),
        mode='markers',
        name=f'{"Alg 1" if id == "OPT" else "ROAR"}',
    ))
    
    figs[2].add_trace(go.Scatter(
        x=temp['lambda'],
        y=temp['Mean Cost'].round(2),
        error_y=dict(array=temp['Std Cost'].round(2)),
        marker=dict(color=Jcolor[id], symbol='circle' if id=='OPT' else 'circle-open', size=5),
        mode='markers',
        name=f'{"Alg 1" if id == "OPT" else "ROAR"} Cost',
    ))

y_title = [r'$\theta_0 \quad Validity$', r"$\theta' \quad Validity$", "Cost"]
for i, fig in enumerate(figs):
    fig.update_xaxes(
        title=dict(
            text='λ',
            font=dict(
                family=font_family,
                color=font_color,
                size=25
            )
            ), 
        showline=True, 
        mirror=True,
        linecolor='black', 
        gridcolor='lightgrey', 
        zerolinewidth=1,
        zerolinecolor='lightgrey',
        )

    fig.update_yaxes(
        title=dict(
            text=y_title[i],
            font=dict(
                family=font_family,
                color=font_color,
                size=25
            ), 
            ), 
        showline=True, 
        mirror=True,
        linecolor='black', 
        gridcolor='lightgrey',
        zerolinewidth=1,
        zerolinecolor='lightgrey',
        )

    fig.update_layout(
        # title=dict(
        #     text=f"{data_names[data]} Dataset", 
        #     x=0.5, 
        #     font=dict(
        #         family=font_family, 
        #         color=font_color, 
        #         size=35
        #         )
        #     ), 
        legend=dict(
            x=0.985,
            y=1.1,
            xanchor='right',
            font=dict(
                family=font_family,
                color=font_color,
                size=15
                ), 
            orientation='h'
            ),
        width=width,
        height=height,
        plot_bgcolor='white',
        paper_bgcolor='white',
        xaxis=dict(
            tickfont=dict(
                family=font_family,
                color=font_color,
                size=20,
            )
        ),
        yaxis=dict(
            tickfont=dict(
                family=font_family,
                color=font_color,
                size=20
            )
        )
        )

    fig.show()
```

```python
names = ['m1_validity', 'wc_validity', 'cost']
for i, fig in enumerate(figs):
    fig.write_image(f'../figs/lambda_effect_{names[i]}_{base_model}_{data}.pdf')
```
